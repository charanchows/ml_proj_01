```python
ds1=ds.drop(['area_type','availability','society'],axis='columns')
ds1.head()
```

```python
ds1.isnull().sum()
```

```python
ds2=ds1.dropna(subset=['location','size','bath'])
```

```python
ds2.isnull().sum()
```

```python
ds3=ds2.fillna(ds2.balcony.median())
```

```python
ds3.isnull().sum()
```

```python
ds3['size'].unique()
```

```python
ds3['BHK']=ds3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
ds3.head()
```

```python
ds3['BHK'].value_counts()
```

```python
ds3[ds3.BHK>23]
```

```python
ds3['total_sqft'].value_counts()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
        
```

```python
ds3[~ds3['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_into_sqft(x):
    s=x.split('-')
    if len(s) == 2:
        return (float(s[0])+float(s[1]))/2
    
    try:
        return float(x)
    except:
        return None
            
```

```python
ds3.total_sqft=ds3.total_sqft.apply(convert_into_sqft)
```

```python
ds3['total_sqft'].unique()
```

```python
ds4=ds3[ds3.total_sqft.notnull()]
```

```python
ds4.head(10)
```

```python
ds4.isnull().sum()
```

```python
ds5=ds4.copy()
ds5['price_per_sqft']=ds4['price']*100000/ds4['total_sqft']
ds5.head(10)
```

```python
ds5.location = ds5.location.apply(lambda x: x.strip())
location_stats=ds5.location.value_counts()
location_stats
```

```python
len(location_stats[location_stats>15])
```

```python
location_stats_10=location_stats[location_stats<15]
```

```python
 ds5.location=ds5.location.apply(lambda x: 'other' if x in location_stats_10 else x )
```

```python
ds5.head(10)
```

```python
ds6=ds5.drop(['total_sqft','price','size'],axis=1)
ds6.head(10)
```
